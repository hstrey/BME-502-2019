```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exon_data = pd.read_csv("exon_length.txt")
exon_data['length'] = exon_data['Exon Chr End (bp)']-exon_data['Exon Chr Start (bp)']+1
```

```python
exon_data['log10']=np.log10(exon_data['length'])
```

```python
plt.hist(exon_data['log10'],bins=100)
plt.xlim((1,4))
```
